```python
# Номер 2.6
vectorCount = 100
angle90Count = 0
angle30Count = 0

V = np.random.random(size = (vectorCount, 2)) - 0.5
q = np.array([2, 1])

V = np.dot(V, q) / np.linalg.norm(q) / np.linalg.norm(V , axis = 1)

angle90Count = (np.arccos(V) * 180 / np.pi < 90).sum() / vectorCount
angle30Count = (np.arccos(V) * 180 / np.pi < 30).sum() / vectorCount

print('Пропорция векторов, для которых угол < 90 = ', angle90Count)
print('Пропорция векторов, для которых угол < 30 = ', angle30Count)
```

```python
# Номер 2.4
def transformation_plot(points, matr):
    
    fig, ax = plt.subplots(1, 2)
    x = points[:, 0]
    y = points[:, 1]
    
    def f(x, y):
        return x + y
    
    ax[0].scatter(x, y, 45, c = f(x, y))
    
    fin = np.dot(points, matr.T)
    x = fin[:, 0]
    y = fin[:, 1]
    
    ax[1].scatter(x, y, 45, c = f(x, y))
    
transformation_plot(np.random.random(size = (150, 2)), np.array([[5, 4], [6, 3]]))
```

```python
# Номер 2.5
fig, ax = plt.subplots(figsize = (12, 6))
x = np.linspace(0, 10, 500)
y = np.sin(x)

x1 = x[0:-1] - x[1:]
y1 = np.array(y[0:-1]) - np.array(y[1:])

anx = (x[0:-1] + x[1:])/2
any = np.divide(y1, x1)

ax.plot(x, np.cos(x))
ax.plot(anx, any)
```

```python
# Номер 1.1
import json
import csv

headers = ['item', 'country', 'year', 'sales']

with open('sales.json', 'r') as file:
    dataJson = json.load(file)

with open('sales.csv', 'w') as csvFile:
    newCsvFile = csv.DictWriter(csvFile, fieldnames = headers)
    newCsvFile.writeheader()

    for item in dataJson:
        countries = item['sales_by_country']

        for country in countries:
            years = countries[country]

            for year in years:
                newCsvFile.writerow({'item':item['item'],'country':country,'year':year,'sales':item['sales_by_country'][country][year]})
```
